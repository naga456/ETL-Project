```python
movie_metadata = os.path.join(file_path, "movie_metadata.csv")
print(movie_metadata)
```

```python
movie_metadata_df = pd.read_csv(movie_metadata)
movie_metadata_df.head()
```

```python
columns = movie_metadata_df.columns
for x in columns:
    print (x)
```

```python
movie_clean1_df = movie_metadata_df[[\
    "movie_title", "plot_keywords", "language", "country", "title_year", "genres", "duration",\
    "director_name", "imdb_score", "movie_imdb_link", "num_critic_for_reviews",\
    "actor_1_name", "actor_2_name", "actor_3_name", "budget", "gross"]].copy()
movie_clean1_df.head()
```

```python
movie_clean1_df.columns = [
    "Movie_Title", "Plot_Keywords", "Language", "Country", "Year_of_Release(year)", "Genres", "Duration(minutes)",\
    "Director_Name", "IMDB_Score", "IMDB_Link", "Number_of_Critic_Reviews",\
    "Actor_1", "Actor_2", "Actor_3", "Film_Budget", "Gross_Income"
    ]
movie_clean1_df.head()
```

```python
print(movie_clean1_df.shape[0])
print(movie_clean1_df.shape[1])
```

```python
movie_clean1_df.dtypes
```

```python
print(movie_clean1_df["Year_of_Release(year)"].max())
print(movie_clean1_df["Year_of_Release(year)"].min())
```

```python
important_list = ["Film_Budget", "Plot_Keywords", "Gross_Income", "Movie_Title", "Director_Name", "Year_of_Release(year)",\
                 "Duration(minutes)", "IMDB_Score", "IMDB_Link"]

movie_clean1_df.dropna(subset=important_list,\
                                         how="any", inplace=True)
print(movie_clean1_df.shape[0])
print(movie_clean1_df.shape[1])
```

```python
movie_clean1_df.head()
```

```python
movie_clean1_df["Genres"] = movie_clean1_df["Genres"].apply(lambda x: x.split('|'))
movie_clean1_df["Plot_Keywords"] = movie_clean1_df["Plot_Keywords"].apply(lambda y: y.split('|'))
movie_clean1_df.head()
```

```python
movie_clean1_df.head()
```

```python
movie_clean1_df.dtypes
```

```python
movie_clean1_df["Apparent_Net_Income"] = movie_clean1_df["Gross_Income"] - movie_clean1_df["Film_Budget"]
movie_clean1_df.head()
```

```python
movie_clean1_df.round(0)

movie_clean1_df["Year_of_Release(year)"] = movie_clean1_df["Year_of_Release(year)"].map('{:.0f}'.format)
movie_clean1_df["Duration(minutes)"] = movie_clean1_df["Duration(minutes)"].map('{:.0f}'.format)
movie_clean1_df["Film_Budget"] = movie_clean1_df["Film_Budget"].map('{:.0f}'.format)
movie_clean1_df["Gross_Income"] = movie_clean1_df["Gross_Income"].map('{:.0f}'.format)
movie_clean1_df["Apparent_Net_Income"] = movie_clean1_df["Apparent_Net_Income"].map('{:.0f}'.format)
```

```python
movie_clean1_df.dtypes
```

```python
movie_clean1_df.head()
```

```python
movie_metadata_clean = os.path.join(file_path, "movie_metadata_clean.csv")
movie_clean1_df.to_csv(movie_metadata_clean)
```
